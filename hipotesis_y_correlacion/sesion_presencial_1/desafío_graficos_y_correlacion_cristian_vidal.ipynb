## 2. Refactor gráficos matplotlib a seaborn

A continuación se presenta una serie de gráficos construídos con matplotlib . Se le pide
refactorizarlos utilizando seaborn .
* Se presenta la función que se utilizó para construírlos. Intente llegar al resultado con mayor similitud.
* Comente los principales resultados de los gráficos

```python
def binarize_histogram(dataframe, variable):
    tmp = dataframe
    tmp['binarize'] = np.where(tmp[variable] > np.mean(tmp[variable]), 1, 0)
    hist_1 = tmp[tmp['binarize'] == 1][variable].dropna()
    hist_0 = tmp[tmp['binarize'] == 0][variable].dropna()
    plt.subplot(1, 2, 1)
    plt.hist(hist_0, alpha=.6, color='lightgrey')
    plt.axvline(np.mean(hist_0))
    plt.title("{0} <= {1}".format(variable, round(np.mean(hist_0), 3)))
    plt.subplot(1, 2, 2)
    plt.hist(hist_1, alpha=.6, color='lightgrey')
    plt.axvline(np.mean(hist_1))
    plt.title("{0} >= {1}".format(variable, round(np.mean(hist_0), 3)))
```

```python
binarize_histogram(df, 'adfert')
```

```python

def binarize_histogram_seaborn(dataframe, variable):
    tmp = dataframe
    col_binarize = f"binarize_{variable}"
    tmp[col_binarize] = np.where(tmp[variable] > np.mean(tmp[variable]), 1, 0)
    hist_0 = tmp[tmp[col_binarize] == 0][variable].dropna()
    hist_1 = tmp[tmp[col_binarize] == 1][variable].dropna()
    
    plt.subplot(1, 2, 1)
    sns.distplot(
        hist_0, 
        rug=True,
        axlabel=variable,
        fit=stats.norm,
        color='sienna').set_title('{0} <= {1}'.format(variable, round(np.mean(hist_0), 3)));
    plt.axvline(round(np.mean(hist_0), 3), color='b', linestyle='--')
    
    plt.subplot(1, 2, 2)
    sns.distplot(
        hist_1, 
        rug=True,
        axlabel=variable,
        fit=stats.norm,
        color='sienna').set_title('{0} >= {1}'.format(variable, round(np.mean(hist_0), 3)));
    plt.axvline(round(np.mean(hist_1), 3), color='b', linestyle='--')
```

```python
binarize_histogram_seaborn(df, 'adfert')
```

Al visualizar los histogramas, observamos que las distribuciones son levemente sesgada hacia la izquierda con una media de 21.88.

```python
def grouped_boxplot(dataframe, variable, group_by):
    tmp = dataframe
    stratify_by = tmp[group_by].unique()
    if len(stratify_by) / 2 > 3:
        fig, ax = plt.subplots(2, len(stratify_by),sharey=True)
    else:
        fig, ax = plt.subplots(1, len(stratify_by),sharey=True)
    for i, n in enumerate(stratify_by):
        ax[i].boxplot(tmp[tmp[group_by] == n][variable])
        ax[i].set_title(n)
```

```python
grouped_boxplot(df, 'adfert', 'region')
```

```python
def grouped_boxplot_seaborn(dataframe, variable, group_by):
    palette='Paired'
    sns.catplot(x=group_by, y=variable, data=dataframe, height=6, kind="box", palette=palette)
```

```python
grouped_boxplot_seaborn(df, 'adfert', 'region')
```

A simple vista, podemos observar lo siguiente:
* El continente africano tiene una mayor fertilidad.
* El continente europeo tiene una distribución mucho mas concentrada.
* La distribución es mas simétrica para el continente americano y europeo.
* Se pueden observar valores atípicos para los continentes: Asia, Europa y Oceanía.

```python
def grouped_scatterplot(dataframe, x, y, group_by):
    tmp = dataframe
    stratify_by = tmp[group_by].unique()
    if len(stratify_by) / 2 > 3:
        fig, ax = plt.subplots(2, len(stratify_by),sharey=True)
    else:
        fig, ax = plt.subplots(1, len(stratify_by),sharey=True)
    for i, n in enumerate(stratify_by):
        tmp_group_plt = tmp[tmp[group_by] == n]
        ax[i].plot(tmp_group_plt[x], tmp_group_plt[y], 'o')
        ax[i].set_title(n)
```

```python
grouped_scatterplot(df, 'school', 'adfert', 'region')
```

```python
def grouped_scatterplot_seaborn(dataframe, x, y, group_by):
    sns.lmplot(x=x, y=y, col=group_by, data=dataframe, fit_reg=True, height=4, col_wrap=3)
```

```python
grouped_scatterplot_seaborn(df,"school", "adfert","region")
```

A simple vista, podemos observar que los gráficos nos enseñan que en la medida que el "promedio de años de escolaridad" va aumentando en la escala, la "Fertilidad adoloscente" disminuye (relación inversamente proporcional)

## 3. Genere un heatmap entre todas las variables

* En base a las variables de interés asignadas, comente cuáles son las principales correlaciones existentes, tomando como criterio de corte aquellas superior a .6

```python
df_subset = df.loc[:, ['country','region','gdp','school','adfert','chldmort','life','pop','urban','femlab','literacy','co2','gini']]
corr_mat = df_subset.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr_mat, annot=True, vmin=.6, robust=True)
```

A simple vista podemos observar un par de patrones claros en las correlaciones (tomando como criterio de corte aquellas superior a .6):

* Hay una importante correlación entre la variable "Tasa de alfabetismo." y "Promedio años de escolaridad" (0.85)
* Hay una importante correlación entre la variable "Toneladas de Co2 mitidas per cápita" y "Producto Interno Bruto per cápita, precios 2005" (0.83)
* Hay una importante correlación entre la variable "Probabilidad de muerte antes de los 5 años por cada 1000" y "Fertilidad adolescente" (0.78)

<b>Al parecer la relación más importante es entre las variables "Tasa de alfabetismo." y "Promedio años de escolaridad" con una correlación de pearson de 0.85</b>.

## 4. En base a las principales correlaciones, sepárelas en un nuevo objeto y calcule la matriz de correlaciones para todas las regiones

* tip: Genere una nueva tabla segmentando con la siguiente sintáxis: tmp = df.loc[:,
['variables', 'a', 'agregar']] . No olvide agregar la variable region .
* tip: Genere un loop para recorrer cada región y generar un heatmap.
* Comente brevemente las principales correlaciones a través de las regiones.

```python
df_subset_2 = df.loc[:, ['co2','gdp','literacy','school','chldmort','adfert','region']]

for region in df['region'].unique():
    df_subset_2['region_bin'] = np.where(df_subset_2['region'] == region, 1, 0)
    new_df = df_subset_2[df_subset_2['region_bin'] == 1].dropna()
    corr_mat = new_df.corr()
    plt.figure(figsize=(10,4))
    ax = plt.axes()
    sns.heatmap(corr_mat, annot=True, vmin=.6, ax=ax, robust=True)
    ax.set_title(region)    
    plt.show()
```

Comentarios
* Para el continente africano la relación más importante es entre las variables "Toneladas de Co2 mitidas per cápita" y "Producto Interno Bruto per cápita" con una correlación de pearson de 0.91.
* Para el continente americano la relación más importante es entre las variables "Promedio años de escolaridad" y "Tasa de alfabetismo" con una correlación de pearson de 0.77.
* Para el continente asiatico la relación más importante es entre las variables "Toneladas de Co2 mitidas per cápita" y " Producto Interno Bruto per cápita" con una correlación de pearson de 0.92.
* Para el continente europeo la relación más importante es entre las variables " Fertilidad adolescente" y "Probabilidad de muerte antes de los 5 años por cada 1000" con una correlación de pearson de 0.69.
* Para el continente de oceanía la relación más importante es entre las variables "Tasa de alfabetismo" y "Promedio años de escolaridad." con una correlación de pearson de 0.98.
